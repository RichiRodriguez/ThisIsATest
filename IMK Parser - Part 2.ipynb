```python
import json, os, glob, csv

files_only = [f for f in glob.glob(input_file_path) if not os.path.isdir(f)]
files_only = sorted(files_only, key=os.path.getctime)[-8:]

offset = 0 if control_was_the_first_block_shown else 20
output = []

for (subblock, file) in enumerate(files_only, 1):
    with open(file) as json_file:
        data = json.load(json_file)

    data = [x for x in data['markCollections'][0]['marks'] if "ExtraData" in x.keys()]
    data = data[1:]
    data = [json.loads(x['ExtraData']) for x in data]
    
    for (i, trial_item) in enumerate(data, 1):
        trial_number = ((subblock - 1) * 5) + i
        scenario = int(trial_item['Vantage_Point'])
        subject_answer_number = answer_set[scenario]['answers'].index(trial_item['Response']) + 1
        correct_answer_number = answer_set[scenario]['answers'].index(correct_answers[scenario]) + 1
        output.append({
            "PartID": partID,
            "Modality": "Control" if trial_number <= 20 else "ISA",
            "Counter Balance Group": "C" if control_was_the_first_block_shown else "I",
            "Subblock": subblock if subblock <= 4 else (subblock - 4),
            "Trial Order": (trial_number - 20) if trial_number > 20 else trial_number,
            "Scenario": scenario,
            "Subject Answered": trial_item['Response'],
            "Time": round(float(trial_item['Elapsed_Seconds']), 2),
            "Correct": 1 if trial_item['Response'] == correct_answers[scenario] else 0,
            "Subject Answer Number": subject_answer_number,
            "Correct Answer Number": correct_answer_number
        })
```

### CSV Export
